```python
import joblib
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python

# Cargar el modelo entrenado y el archivo de test
modelo = joblib.load('../Modelos/best_stacking_model_3_PRUEBA.joblib')
mappings = joblib.load('label_encodings_mappings_2.joblib')
test_prueba = pd.read_csv('../Test/TEST_PRUEBAS_2.csv')
```

```python



# Definir el mapeo de resultados
mapeo_resultados = {0: '1', 1: '2'}

# Obtener las probabilidades de predicci√≥n
resultado_proba = modelo.predict_proba(test_prueba)
resultado_proba_porcentaje = np.round(resultado_proba * 100, 2)
# Obtener las predicciones de clase
predicciones_clase = modelo.predict(test_prueba)

# Si predicciones_clase contiene elementos que son arrays, extraer el primer elemento
if isinstance(predicciones_clase[0], (list, np.ndarray)):
    predicciones_clase = [clase[0] for clase in predicciones_clase]

# Mapear las predicciones de clase a los resultados
predicciones_resultado = [mapeo_resultados[clase] for clase in predicciones_clase]

# Crear un DataFrame con las probabilidades y los resultados mapeados
probabilidades_df = pd.DataFrame(resultado_proba_porcentaje, columns=['Proba_1', 'Proba_2'])
probabilidades_df['Resultado'] = predicciones_resultado

# Mostrar el DataFrame resultante
print(probabilidades_df)


```

```python
probabilidades_df["Jornada"] = test_prueba ["Jornada"]
probabilidades_df["Local"] = test_prueba["Local"]
probabilidades_df["Visitante"] = test_prueba["Visitante"]
```

```python
mappings_local = mappings['Local']
mappings_visitante = mappings['Visitante'] 
inverse_mappings_local = {v: k for k, v in mappings_local.items()}
inverse_mappings_visitante = {v: k for k, v in mappings_visitante.items()}
```

```python
probabilidades_df['Local'] = probabilidades_df['Local'].map(inverse_mappings_local)
probabilidades_df['Visitante'] = probabilidades_df['Visitante'].map(inverse_mappings_visitante)
```
